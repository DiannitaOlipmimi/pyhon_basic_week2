## 6.2. Strings dan List
---
Strings merupakan kumpulan karakter yang berurutan dari kiri ke kanan. strings tanpa karakter disebut dengan *empty string* dan biasa diwakili degan '' atau "". sedangkan List adalah kumpulan nilai berurutan yang dapat memiliki macam-macam tipe data (tidak hanya strings), dimana setiap nilainya diidentifikasi dengan index dan nilainya disebut elemen.

```python
s = 'Hello world'
print(type(s))
```

```python
t = [10, 20, 30, 40]
u = ["spam", "bungee", "swallow"]

print(type(u), type(t))
```

```python
a = ["hello", 2.0, 5, [10, 20]]

print(a[1]) #index 1 = elemen list ke 2
```

## Tuples
---
merupakan urutan nilai yang dapat terdiri dari tipe data apapun seperti list namun urutannya dipisahkan oleh koma (,) dan diapit oleh tanda kurung (list menggunakan []). selain perbedaan tanda yang mengapit, tuple tidak dapat diubah setelah dibuat sedangkan list dapat diubah.

```python
julia = ("Julia", "Roberts", 1967, "Duplicity", 2009, "Actress", "Atlanta, Georgia")

print(type(julia))
```

```python
t = (5,)
print(type(t))

x = (5)
print(type(x))
```
