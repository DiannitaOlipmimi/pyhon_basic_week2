## 6.6. The Slice Operator
---
substring dari string disebut dengan slice. slice juga dapat diartikan sebagai mengambil karakter dari suatu string. operator dalam string adalah [n:m] dimana n merupakan awalan string yang diambil sampai karakter sebelum m.

```python
julia = ("julia", "robert", "1967", "duplicity",2009, "actress", "atlanta, geogia")

print(julia[2]) #indexing
print(julia[2:6]) #slicing
print(len(julia)) #banyak karakter

julia=julia[:3]+("eat pray love", 2010)+julia[5:]
print(julia)
```

```python
singers = "Peter, Paul, and Mary"
print(singers[0:5])
print(singers[7:11])
print(singers[17:21])
```

```python
fruit = "banana"
print(fruit[:3]) #slice 3 awal
print(fruit[3:]) #slice 3 akhir
```

```python
fruit[:] #mengambil seluruh karakter
```

## 6.8. Count and Index
---
salah satu *built-in function* yang sering digunakan adalah *count* dan *index*. *function* ini dapat digunakan untuk tipe data string dan list.
---
**Count**
merupakan fungsi dengan 1 argumen yang berisi hal yang akan dihitung. fungsi ini mengembalikan banyaknya argumen yang muncul dalam suatu string atau list. argumen dalam count hanya dapat berisi string.
**Index**
sama seperti fungsi *count*, index memiliki 1 argumen. index mengembalikan index paling kiri dari argumen. index dapat digunakan baik list maupun string.

```python
a = "I have had n apple on my desk before!"
print(a.count("e")) #count e
print(a.count("ha")) #count word ha
```

```python
z = ['atoms', 4, 'neutron', 6, 'proton', 4, 'electron', 4, 'electron', 'atoms']
print(z.count("4")) #count string
print(z.count(4)) #count int
print(z.count("a"))
print(z.count("electron"))
```

## Split dan Join Strings
---
fungsi split memisahkan string menjadi list dengan kata kata. argumen dalam split dapat berupa **delimiter** yang dapat digunakan untuk memisahkan string berdasarkan kata dalam argumen. kebelikan dari split adalah join. join adalah fungsi yang digunakan untuk menyatukan elemen dengan argumen separator yang dapat dipilih.

```python
song = "The rain in Spain..."
wds = song.split() #split by space
print(wds)
```

```python
song = "The rain in Spain..."
wds = song.split('ai') #split by the word ai
print(wds)
```

```python
wds = ["red", "blue", "green"]
glue = ';'
s = glue.join(wds) #menyatukan
print(s)
print(wds)

print("***".join(wds))
print("".join(wds))
```

```python
weather = ["sunny", "cloudy", "partially sunny", "rainy", "storming", "windy", "foggy", "snowy", "hailing"]

for condition in weather :
    print("the word is", len(condition), "characters")
    first_char = condition[0]
    last_char = condition[-1]
    
    print("the first char is:" + first_char)
    print("the last char is:" + last_char)
```
