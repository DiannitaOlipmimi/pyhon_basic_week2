```python
fruit = "Banana"
print(len(fruit))
```

## 6.6. The Slice Operator
---
substring dari string disebut dengan slice. slice juga dapat diartikan sebagai mengambil karakter dari suatu string. operator dalam string adalah [n:m] dimana n merupakan awalan string yang diambil sampai karakter sebelum m.

```python
julia = ("julia", "robert", "1967", "duplicity",2009, "actress", "atlanta, geogia")

print(julia[2]) #indexing
print(julia[2:6]) #slicing
print(len(julia)) #banyak karakter

julia=julia[:3]+("eat pray love", 2010)+julia[5:]
print(julia)
```

```python
singers = "Peter, Paul, and Mary"
print(singers[0:5])
print(singers[7:11])
print(singers[17:21])
```

```python
fruit = "banana"
print(fruit[:3]) #slice 3 awal
print(fruit[3:]) #slice 3 akhir
```

```python
fruit[:] #mengambil seluruh karakter
```

```python
#slice dalam list
a_list = ['a', 'b', 'c', 'd', 'e', 'f']
print(a_list[1:3])
print(a_list[:4])
print(a_list[3:])
print(a_list[:])
```

```python
#slice dalam tuple
julia = ("Julia", "Roberts", 1967, "Duplicity", 2009, "Actress", "Atlanta, Georgia")
print(julia[2])
print(julia[2:6])

print(len(julia))

julia = julia[:3] + ("Eat Pray Love", 2010) + julia[5:]
print(julia)
```

```python
s = "python rocks"
print(s[3:8])
```

```python
alist = [3, 67, "cat", [56, 57, "dog"], [ ], 3.14, False]
print(alist[4:])
```

```python
L = [0.34, '6', 'SI106', 'Python', -2]
print(len(L[1:-1]))
```

## 6.7. Concatenation and Repetition
---
penggabungan dan perulangan. dalam jenis data String penggabungannya dapat dilajukan dengan menggunakan operator (+). hal ini juga berlaku pada operator-operator lain seperti (*) dimana operator tersebut akan melakukan perulangan sebanyak n kali.

```python
fruit = ["apple","orange","banana","cherry"]
print([1,2] + [3,4])
print(fruit+[6,7,8,9])

print([0] * 4)
```

```python
alist = [1,3,5]
blist = [2,4,6]
print(alist + blist)
```

```python
alist = [1,3,5]
print(alist * 3)
```
