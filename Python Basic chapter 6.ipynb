## 6.2. Strings dan List
---
Strings merupakan kumpulan karakter yang berurutan dari kiri ke kanan. strings tanpa karakter disebut dengan *empty string* dan biasa diwakili degan '' atau "". sedangkan List adalah kumpulan nilai berurutan yang dapat memiliki macam-macam tipe data (tidak hanya strings), dimana setiap nilainya diidentifikasi dengan index dan nilainya disebut elemen.

```python
s = 'Hello world'
print(type(s))
```

```python
t = [10, 20, 30, 40]
u = ["spam", "bungee", "swallow"]

print(type(u), type(t))
```

```python
a = ["hello", 2.0, 5, [10, 20]]

print(a[1]) #index 1 = elemen list ke 2
```

## Tuples
---
merupakan urutan nilai yang dapat terdiri dari tipe data apapun seperti list namun urutannya dipisahkan oleh koma (,) dan diapit oleh tanda kurung (list menggunakan []). selain perbedaan tanda yang mengapit, tuple tidak dapat diubah setelah dibuat sedangkan list dapat diubah.

```python
julia = ("Julia", "Roberts", 1967, "Duplicity", 2009, "Actress", "Atlanta, Georgia")

print(type(julia))
```

```python
t = (5,)
print(type(t))

x = (5)
print(type(x))
```

## 6.3. Index Operator: Strings, List, Tuple
---
indeks operator dalam Python menggunakan tanda kurung siku ([]) untuk memilih satu karakter dalam string. karakter dalam string sendiri diakses melalui posisi atau nilai indeksnya dan dimulai dari 0.

```python
school = "Luther College"
m = school[2] #indeks ke 3
print(m)

lastchar = school[-1] #indeks ke 14
print(lastchar)
```

```python
numbers = [17, 123, 87, 34, 66, 8398, 44]
print(numbers[2])
print(numbers[9-8])
print(numbers[-2])
```

```python
prices = (1.99, 2.00, 5.50, 20.95, 100.98)
print(prices[0])
print(prices[-1])
print(prices[3-5])
```

```python
s = "python rocks"
print(s[3])
```

```python
s = "python rocks"
print(s[2] + s[-4])
```

```python
alist = [3, 67, "cat", [56, 57, "dog"], [ ], 3.14, False]
print(alist[5])
```
